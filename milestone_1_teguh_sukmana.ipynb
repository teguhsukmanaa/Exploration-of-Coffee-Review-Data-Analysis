# Analisis dan Perhitungan

### Analisis Rating Kopi

Untuk melakukan analisis tren kopi berdasarkan dataset review kopi ini, kita bisa mengawalinya dengan mengelompokkan rerata rating produk kopi berdasarkan tahunnya.

```python
# menghitung rerata rating dari tiap nama kopi dan mengelompokkan berdasarkan tahunnya 
avg_rating_year = df.groupby(['review_year', 'name'])['rating'].mean().reset_index()
avg_rating_year
```

Diperoleh dataframe 'avg_rating_year' yang berisi **rerata rating untuk setiap produk kopi dan dikelompokkan berdasarkan tahunnya**. Melalui dataframe ini, kita bisa menganalisis produk kopi berdasarkan rating tertentu dan pada periode waktu tertentu pula.

```python
# Mencari/memilih produk kopi dengan rating tertinggi setiap tahunnya
highest_rated_year = avg_rating_year.groupby('review_year')['rating'].max().reset_index()
highest_rated_coffee = pd.merge(avg_rating_year, highest_rated_year, on=['review_year', 'rating'])

highest_rated_coffee
```

Berikut diperoleh dataframe 'highest_rated_coffee' sebagai hasil temuan untuk **produk-produk kopi dengan rating tertinggi pada tiap tahun**. Selain itu, terlihat pula bahwa terdapat lebih dari 1 produk kopi yang mendapat rating tertinggi di beberapa tahun seperti tahun 2018, 2019, dan 2020. Dataframe ini dapat digunakan jika kita ingin menganalisis faktor pada produk dengan rating tertinggi saja.

### Analisis Korelasi Harga dan Rating Kopi


Kira-kira berapa harga dari produk-produk kopi yang memiliki rating tertinggi?

```python
hr_coffee_price = pd.merge(highest_rated_coffee, df[['name', 'review_year', '100g_USD']], on=['name', 'review_year'], how='left')
hr_coffee_price
```

```python
# melihat distribusi data dari kolom '100g_USD' atau harga pada produk kopi dengan rate tinggi
print("Mean:", round(hr_coffee_price['100g_USD'].mean(), 2)) # menampilkan rata-rata dan membulatkannya
print("Median:", hr_coffee_price['100g_USD'].median()) # menampilkan median/nilai tengah
print("Standar Deviasi:", round(hr_coffee_price['100g_USD'].std(), 3)) # menampilkan rata-rata dan membulatkannya
print("Skewness:", hr_coffee_price['100g_USD'].skew()) # menampilkan nilai skewness
print("Kurtosis:", hr_coffee_price['100g_USD'].kurtosis()) # menampilkan nilai kurtosis
```

Uji statistik deskriptif menunjukkan bahwa produk-produk kopi dengan rating tertinggi ini memiliki rerata harga yang berkisar pada 16.67 USD per 100gr dengan nilai tengah harganya yang berada di 15.43 USD. Kemudian berdasarkan distribusi datanya yang dilihat melalui nilai std, skewness, dan kurtosis, dapat diartikan bahwa data harga dari kelompok kopi dengan rating tertinggi ini memiliki data yang terdistribusi mendekati normal dengan kemungkinan outlier yang cendrung rendah. Sehingga berdasarkan analisis ini, kita dapat **berasumsi** bahwa **produk kopi dengan rating tertinggi ternyata memiliki harga yang tidak begitu mahal atau memiliki harga yang tergolong menengah**. <br>
<br>Namun apakah benar demikian? bukankah biasanya produk dengan harga yang mahal memiliki kualitas yang baik sehingga memiliki rating penilaian yang tinggi? Untuk itu kita dapat menganalisis bagaimana korelasi antara harga dan rating yang dimiliki oleh produk kopi.

Sebelum melakukan analisis korelasi antara data kolom harga dan rating kopi, terlebih dahulu dilakukan analisis distribusi data pada kolom harga kopi di dataframe utama

```python
# melihat distribusi data dari kolom '100g_USD' atau harga kopi secara keseluruhan
print("Mean:", round(df['100g_USD'].mean(), 2)) # menampilkan rata-rata dan membulatkannya
print("Median:", df['100g_USD'].median()) # menampilkan median/nilai tengah
print("Standar Deviasi:", round(df['100g_USD'].std(), 3)) # menampilkan rata-rata dan membulatkannya
print("Skewness:", df['100g_USD'].skew()) # menampilkan nilai skewness
print("Kurtosis:", df['100g_USD'].kurtosis()) # menampilkan nilai kurtosis
```

Berdasarkan statistik deskriptif dari kolom '100g_USD' atau harga kopi secara keseluruhan, diketahui bahwa rerata harga kopi dari tahun 2017-2022 berkisar pada 10 USD dengan nilai tengahnya di harga 6.17 USD. Jika hanya melihat 2 nilai tersebut maka terlihat harga kopi pada 6 tahun review tergolong murah.Namun, yang perlu diperhatikan adalah pada nilai standar deviasi sebesar 11.966, nilai skewness sebesar 5.318001774797196, dan nilai kurtosis sebesar 38.34749340531837 pada hasil perhitungan tesebut. Ketiga nilai tersebut tergolong tinggi untuk sebuah persebaran data, atau yang berarti bahwa terdapat jarak persebaran data yang jauh/luas (data tidak terdistribusi normal) serta memiliki outlier/extreme value didalamnya.
<br><br>Berdasarkan hal tersebut, koefisien korelasi hubungan antara harga dan rating yang akan dilakukan mungkin tidak akan begitu akurat.

Digunakan analisis korelasi Spearman karena kedua data yang akan diuji merupakan data continous dan terindikasi tidak terdistribusi normal.

```python
# Spearman Correlation Test
corr_price_rate = df['100g_USD'].corr(df['rating'], method='spearman')
print("rho-correlation", corr_price_rate)
```

Hasil dari analisis korelasi adalah 0.3028603241852044. Nilai yang positif dari analisis mengindikasikan **hubungan monotonik positif** antara kedua variabel, atau berarti ketika harga kopi meningkat, maka rating juga cendrung meningkat. Namun, besaran nilai sebesar 0.302 mengindikasikan bahwa korelasi diantara harga dan rating relatif moderat/sedang. Meskipun terdapat hubungan positif antara harga dan rating, bukan barati menaikkan harga kopi secara langsung akan menghasilkan peringkat yang lebih tinggi karena analisis korelasi Spearman tidak menyiratkan hubungan sebab-akibat.

```python
from matplotlib import pyplot as plt
import seaborn as sns

# Memvisualisasikan hubungan harga dan rating menggunakan scatter plot dengan garis regresi
plt.figure(figsize=(8, 6))
sns.regplot(x='100g_USD', y='rating', data=df, scatter_kws={'alpha':0.5})
plt.title('Price vs Rating Correlation')
plt.xlabel('100g_USD')
plt.ylabel('Rating')
plt.grid(True)
plt.show()
```

Terlihat bahwa garis regresi menunjukkan hubungan yang positif, namun dapat terlihat pula bahwa terdapat kecendrungan data yang berkumpul pada bagian sisi kiri-atas yang mengindikasikan bahwa cukup banyak data yang memiliki nilai rating tinggi dengan harga yang terjangkau, sehingga dapat menjadi gambaran pula bagi perusahaan untuk menjadikan kopi yang berapa pada kuadran tersebut sebagai pilihan alternatif.

### Analisis Korelasi Roast Level Rating Kopi

Faktor lain yang diduga memberi pengaruh pada rating kopi adalah roast level dari produk kopi, kira-kira jenis roast level apa yang dimiliki oleh produk-produk kopi dengan rating tertinggi tersebut?

```python
hr_coffee_roast = pd.merge(highest_rated_coffee, df[['name', 'review_year', '100g_USD', 'roast']], on=['name', 'review_year'], how='left')
hr_coffee_roast
```

Jika melihat roast level pada kopi-kopi rating tertinggi di tiap tahun, dapat terlihat bahwa rating tertinggi tiap tahun didominasi oleh roast level medium-light dan light. Apakah itu berarti level medium-light lebih baik atau hanya kebetulan menempati posisi rating tertinggi?

Untuk menganalisis keterkaitan roast level dengan rating kopi yang didapat, kita dapat menganalisisnya pada dataframe utama.

```python
roast_rating = df.groupby('roast')['rating'].mean().reset_index()
roast_rating
```

Jika di lakukan rerata rating maka sekilas terlihat bahwa perbedaan nilai rating diantara roast level tidak terlalu jauh. Untuk menganalisis hal ini kita dapat menggunakan **Uji ANOVA** dalam melihat signifikansi perbandingan rerata tiap kelompok roast level tersebut.<br>
<br>
Berikut hipotesis yang diajukan:
<br>H0 = Tidak ada perbedaan nilai rating yang signifikan di antara 5 roast level
<br>H1 = Terdapat perbedaan nilai rating yang signifikan di antara 5 roast level

```python
from scipy import stats

anova_roast_rate = stats.f_oneway(*(df[df['roast'] == roast]['rating'] for roast in df['roast'].unique()))
print('p-value:',anova_roast_rate.pvalue)
```

Berdasarkan uji ANOVA, diperoleh p-value sebesar 1.3533124282330887e-11 atau jauh lebih kecil dibandingkan significance level (p-value < 0.05), yang mana hal ini berarti H0 ditolak, atau dapat dikatakan bahwa **terdapat perbedaan yang signifikan secara statistik diantara rerata rating tiap roast level**.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='roast', y='rating', data=df)
plt.title('Ratings by Roasting Level Comparison')
plt.xlabel('Roasting Level')
plt.ylabel('Rating')
plt.show()
```

Jika divisualisasikan, dapat terlihat bahwa persebaran data dari tiap kelompok roast level tidak berdistribusi normal, sehingga kurang tepat jika menggunakan ANOVA.

### Analisa Pola Tren Kopi

Jika sebelumnya kita sudah menganalisa kopi-kopi yang menempati posisi rating tertinggi di tiap tahunnya, berikutnya kita akan menganalisis pola tren yang terjadi dari dataset ini.

Namun, untuk menganalisa pola tren kopi, kita tidak bisa hanya menggunakan produk kopi yang mmeiliki nilai tertinggi saja karena menurut sumber acuan (https://www.coffeereview.com/interpret-coffee/) terdapat rentang nilai rating yang menginterpretasikan level tertentu, yang mana nilai rating 95-100 tergolong pada level 'exceptional' atau hits(sangat enak). Berdasarkan hal tersebut, kita akan menganalisis pola tren seperti roast level apa yang mendominasi dan karakteristik rasa apa yang mendominasi berdasarkan data produk yang memiliki rating 95-100.

```python
upper_rate_coffee = df[(df['rating'] >= 95) & (df['rating'] <= 100)]
upper_rate_coffee = upper_rate_coffee.reset_index()
upper_rate_coffee
```

```python
import matplotlib.pyplot as plt

# menghitung jumlah tiap roast level
roast_counts = df['roast'].value_counts()

# membuat pie chart
plt.figure(figsize=(8, 8))
plt.pie(roast_counts, labels=roast_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Roast Levels Proportion')
plt.axis('equal')
plt.show()
```

Dapat diketahui bahwa yang mendominasi adalah medium-light level
